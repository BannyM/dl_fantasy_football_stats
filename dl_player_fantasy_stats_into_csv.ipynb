Web Scraping NFL data project goal is to create a .csv file of relevant NFL stats for the 2021 Season

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
year = 2021
url = "https://www.pro-football-reference.com/years/{}/fantasy.htm".format(year)
html = urlopen(url)
soup = BeautifulSoup(html)
categories = [th.getText() for th in soup.findAll('tr')[1].findAll('th',class_= "poptip")]
categories = categories[1:26]
```

```python
table_body = soup.findAll('tr', class_ = lambda table_rows: table_rows != "thead")
table_data = [[td.getText() for td in table_body[i].findAll('td')]
                for i in range(len(table_body))]
for data in table_data:
    del data[-7:]
table_data = table_data[2:]
```

```python
stats = pd.DataFrame(data = table_data, columns = categories)
stats.head()
```

```python
stats = stats.replace(r'', 0, regex=True)
stats.head()
```

```python
stats.to_csv('{}playerstats.csv'.format(year))
```
